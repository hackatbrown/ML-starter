# MatplotLib short introduction
Matplotlib is a powerful library in Python for creating static, animated, and interactive visualizations.

Matplotlib is often used to visualize datasets being fed into PyTorch models, especially when working with images

This guide will help you get started with the basics of matplotlib, like how to create line plots, bar charts, and scatter plots.




Prerequisites

Ensure you have the matplotlib library installed. If not, run:

```python
pip install matplotlib

#or if you are using conda
#conda install matplotlib
```

Importing Matplotlib
To start using matplotlib, import it in your Python script or Jupyter notebook:
Python


```python
import matplotlib.pyplot as plt
```

Creating a Line Plot
A line plot is one of the most basic types of plots:


```python
#Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 16]

#Creating the plot
plt.plot(x, y, marker='o')

#Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')

#Display the plot
plt.show()
```

# Bar charts are useful for comparing categorical data

```python
#Sample data
categories = ['A', 'B', 'C', 'D']
values = [1, 2, 3, 4]

#Creating the bar chart
plt.bar(categories, values, color='green')

#Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

#Display the plot
plt.show()
```

Creating a Scatter Plot


```python
#Sample data
x = [5, 7, 8, 9, 10]
y = [10, 14, 13, 17, 19]

#Creating the scatter plot
plt.scatter(x, y, color='red')

#Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

#Display the plot
plt.show()
```

Once you're comfortable with these basics, you can explore more advanced topics like subplots, histograms, and customizing plot styles. The official [Matplotlib documentation](https://matplotlib.org/stable/contents.html) is a great resource for deep dives.
